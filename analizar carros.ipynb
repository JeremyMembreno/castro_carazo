```python
filename = 'autos.csv' #lugar del archivo o nombre
filename2 = 'Autos1.csv' #lugar del archivo2 o nombre
# --------------- nombre de las columnas ---------------- 
columnas = [
    "symboling_1",  # "1 Número de secuencia"
    "normalized_2",  # "2 Valor desconocido"
    "make_3",  # "3 Marca del vehículo"
    "fuel_4",  # "4 Tipo de combustible"
    "aspirassion_5",  # "5 Tipo de aspiración"
    "doors_6",  # "6 Número de puertas"
    "style_7",  # "7 Tipo de carrocería"
    "wheels_8",  # "8 Tracción del vehículo"
    "engine_9",  # "9 Posición del motor"
    "base_10",  # "10 Distancia entre ejes"
    "length_11",  # "11 Longitud del vehículo"
    "width_12",  # "12 Anchura del vehículo"
    "heingt_13",  # "13 Altura del vehículo"
    "curb_14",  # "14 Peso del vehículo"
    "engineType_15",  # "15 Tipo de motor"
    "cylinder_16",  # "16 Número de cilindros"
    "engineSize_17",  # "17 Desplazamiento del motor"
    "system_18",  # "18 Tipo de sistema de combustible"
    "bore_19",  # "19 Diámetro del cilindro"
    "stroke_20",  # "20 Carrera del cilindro"
    "compress_21",  # "21 Relación de compresión del motor"
    "horsepower_22",  # "22 Potencia neta del motor"
    "peak_23",  # "23 RPM máximas del motor"
    "city_24",  # "24 Consumo de combustible en ciudad"
    "higway_25",  # "25 Consumo de combustible en carretera"
    "price_26"  # "26 Precio del vehículo"
]

# --------------- nombre de las columnas ---------------- 


data = pd.read_csv(filename, names=columnas, index_col = None, sep = ',') #leer el archivo y ponerle las columnas porque no tiene
data2= pd.read_csv(filename2, names=columnas, index_col = None, sep = ',') #leer el archivo y ponerle las columnas porque no tiene

```

<h3 Style ='color:#F2668B'>4. Trabajo con valores faltantes </h3>
<p>Que se hace para el manejo de valores faltantes ? </p>

<ol>
    <li>identificar como se muestran los valores faltantes</li>
    <li>contar la cantidad de valores faltantes en el set de datos</li>

    <li>eliminar filas</li>

<h5 Style ='color:#F2668B'>4.1 identificar como se muestran los valores faltantes </h5>

```python
dfAutos.replace('?', np.nan, inplace = True)
dfAutos.head(5)
```

<h5 Style ='color:#F2668B'>4.2 contar la cantidad de valores faltantes en el set de datos</h5>

```python
dftemp = dfAutos.isnull()
dftemp.head()
```

```python
for column in dftemp.columns.values.tolist():
    print(column)
    print(dftemp[column].value_counts())
    print("")
```

<h5 Style ='color:#F2668B'>4.3 imputacion de valores numericos</h5>

```python
av_normalized_2 = dfAutos['normalized_2'].astype("float").mean(axis=0)
print("la media de normalized_2 es : ", av_normalized_2)

av_bore_19 = dfAutos['bore_19'].astype("float").mean(axis=0)
print("la media de bore_19 es: ", av_bore_19)

av_stroke_20 = dfAutos['stroke_20'].astype("float").mean(axis=0)
print("la media de stroke_20 es: ", av_stroke_20)

av_horsepower_22 = dfAutos['horsepower_22'].astype("float").mean(axis=0)
print("la media de horsepower_22 es: ", av_horsepower_22)

av_peak_23 = dfAutos['peak_23'].astype("float").mean(axis=0)
print("la media de peak_23 es: ", av_peak_23)
```

```python
dfAutos["normalized_2"].replace(np.nan, 122, inplace = True)
dfAutos["bore_19"].replace(np.nan, 3.32, inplace = True)
dfAutos["stroke_20"].replace(np.nan, 3.28, inplace = True)
dfAutos["horsepower_22"].replace(np.nan, 102, inplace = True)
dfAutos["peak_23"].replace(np.nan, 5100, inplace = True)

dfAutos.style
```

<h5 Style ='color:#F2668B'>4.4 imputacion de valores categoricos</h5>

```python
dfAutos['doors_6'].value_counts().idxmax()
```

```python
dfAutos['doors_6'].replace(np.nan, 'four', inplace = True)
dfAutos.head()
```

```python
dfAutos[['bore_19', 'stroke_20']] = dfAutos[['bore_19', 'stroke_20']].astype(float) #este codigo lo corrige
dfAutos.info() # tipo de columnas corregido completamente
```

```python
dfAutos.describe()
```

<h5 style='color:#F2668B'>5.2 convertir valores numerico, en funcion de la interpretacion del medio </h5>

<h5 Style ='color:#1B7F79'>6.    ¿Cuál es margen promedio de ganancia por cada una de las propiedades vendidas?</h5>

la ganancia es la resta entre saleAmount - assessedValue y el promedio total es la suma de todas las ganancias totales entre la cantidad de ventas

```python
data.head()
```

```python
margen_promedio = data[['saleAmount', 'assessedValue']]

# margen_promedio[['saleAmount', 'assessedValue', 'margen_ganancias']] = margen_promedio[['saleAmount', 'assessedValue', 'margen_ganancias']].astype(int
```

```python
margen_promedio['margen_ganancias'] = margen_promedio['saleAmount'] - margen_promedio['assessedValue']
margen_promedio[['saleAmount', 'assessedValue', 'margen_ganancias']] = margen_promedio[['saleAmount', 'assessedValue', 'margen_ganancias']].astype(int)
margen_promedio
```

```python
promedio_ganacia_total = margen_promedio['margen_ganancias'].mean()
print("El promedio de ganancia total para cada una de las propiedades vendidas es de: ", int(promedio_ganacia_total))
```

<h5 Style ='color:#F2668B'>4.5 Eliminar filas con valores faltantes</h5>

```python
valores_faltantes = data[data[['serialNumber','listYear','dateRecorded','town','address','assessedValue','saleAmount','salesRatio']].isnull().any(axis = 1)]
valores_faltantes
```

```python
data.dropna(subset = ['serialNumber','listYear','dateRecorded','town','address','assessedValue','saleAmount','salesRatio'],axis = 0, inplace = True)
# el index sale mal con este codigo se va a resetear y agregar en el index numeros iniciando desde el 1
data = data.reset_index(drop=True)
data.index += 1

# dfAutos,style()
```
